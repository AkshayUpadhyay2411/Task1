```python
import pandas as pd

# Specify the path to the CSV file
csv_file_path = 'sheets/Task1 - Sheet1.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Now you can work with the DataFrame
# For example, you can print the first few rows
print(df.to_string())
```

```python
# Modification using ChatGPT 3
# Step 1: Read the CSV file
csv_data = '''Student Name,Roll Number,Maths,Science,Total
Anuj Kumar,1,10,20,30
Ayush Kumar,,20,30,50
Harsh Bang,3,30,,30
Aryan Kumar,4,40,50,90
Kartik Gupta,,,60,60
Priya Sharma,6,10,70,80
Monu Mishra,7,20,80,100
Akshay Upadhyay,,30,40,70
'''

# Step 2: Parse the CSV data
lines = csv_data.strip().split('\n')
header = lines[0].split(',')
rows = [line.split(',') for line in lines[1:]]

# Step 3: Find the mean of the "Science" column
science_values = [int(row[3]) for row in rows if row[3]]  # Collect non-empty values
print(science_values)

mean_science = sum(science_values) / len(science_values) if science_values else 0  # Calculate mean
print(mean_science)

# Step 4: Update the empty values in the "Science" column with the mean
for row in rows:
    if not row[3]:  # Check for empty value in "Science" column
        row[3] = str(mean_science)

# Step 5: Generate the final CSV string
final_csv = ','.join(header) + '\n' + '\n'.join([','.join(row) for row in rows])

# Step 6: Print the final CSV
print(final_csv)
```
